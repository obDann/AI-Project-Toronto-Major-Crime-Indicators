```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error
import statsmodels.api as sm
import patsy
import datetime
```

```python
mci_data = pd.read_csv("for_python_modelling.csv")
mci_data
```

```python
formula = 'NUM_CRIMES~OCC_YEAR + OCC_MONTH_NORMAL + OCC_DAY_NORMAL + COVID_ERA + IS_FRIDAY + HOLIDAY_DESCRIPTION '
formula += '+ DIVISION*NEIGHBOURHOOD_158*PREMISES_TYPE '
formula += '+ MCI_CATEGORY'

y, X = patsy.dmatrices(formula, mci_data, return_type='dataframe')
```

Getting errors down below, will consider other methods

```python
print(datetime.datetime.now())
i = 0
kf = KFold(n_splits=5, shuffle=True, random_state = 1001346377)

# Store MAE scores for each fold
mae_scores = []

# Cross-validation loop
for train_index, test_index in kf.split(X):
    i += 1
    print("fold", i)
    print("time", datetime.datetime.now())
    # Split data into training and test sets
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit the GLM model (Poisson regression as an example, replace family if necessary)
    model = sm.GLM(y_train, X_train, family=sm.families.Poisson())
    result = model.fit()
    
    # Make predictions on the test set
    y_pred = result.predict(X_test)
    
    # Calculate MAE for the current fold
    mae = mean_absolute_error(y_test, y_pred)
    mae_scores.append(mae)

# Print out the average MAE across all folds
print("Mean MAE across all folds:", np.mean(mae_scores))
print("time", datetime.datetime.now())
```
